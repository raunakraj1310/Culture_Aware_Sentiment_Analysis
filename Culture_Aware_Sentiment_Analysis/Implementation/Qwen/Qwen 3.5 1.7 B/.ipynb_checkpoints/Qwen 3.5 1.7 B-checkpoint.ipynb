```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "Qwen/Qwen3-1.7B"

tokenizer = AutoTokenizer.from_pretrained(model_name)

tokenizer.padding_side = "left"

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto"
)

print("Model loaded")
```

```python
import os
import json
import re
import pandas as pd
import torch
from tqdm.auto import tqdm
from collections import defaultdict

# --------------------------------------------------
# QWEN FIXES
# --------------------------------------------------

tokenizer.padding_side = "left"

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# --------------------------------------------------
# FOLDERS
# --------------------------------------------------

os.makedirs("state_results", exist_ok=True)
os.makedirs("region_results", exist_ok=True)

# --------------------------------------------------
# PROMPT
# --------------------------------------------------

PROMPT_TEMPLATE = """
You are an expert sentiment classifier.

Your task is to classify sentiment by considering:
- cultural context
- regional practices
- social norms
- geographical sensitivity
- local interpretation

A sentence may express different sentiment in different locations.

STRICT RULES:
- Output ONLY valid JSON
- No explanation
- No markdown
- No extra text

SCHEMA:
{{
    "sentiment":"positive | neutral | negative"
}}

Location:
{location}

Text:
{text}

Return ONLY JSON.
"""

# --------------------------------------------------
# STATES
# --------------------------------------------------

LOCATIONS = {
    # "Punjab": "Punjab - Northern Cultural Region",
    # "Haryana": "Haryana - Northern Cultural Region",
    # "Delhi": "Delhi - Northern Cultural Region",
    # "Himachal_Pradesh": "Himachal Pradesh - Northern Cultural Region",
    # "Jammu_and_Kashmir": "Jammu and Kashmir - Northern Cultural Region",
    # "Ladakh": "Ladakh - Northern Cultural Region",
    # "Uttarakhand": "Uttarakhand - Northern Cultural Region",
    # "Uttar_Pradesh": "Uttar Pradesh - Northern Cultural Region",
    # "Chandigarh": "Chandigarh - Northern Cultural Region",

    # "Tamil_Nadu": "Tamil Nadu - Southern Cultural Region",
    # "Karnataka": "Karnataka - Southern Cultural Region",
    # "Kerala": "Kerala - Southern Cultural Region",
    # "Andhra_Pradesh": "Andhra Pradesh - Southern Cultural Region",
    # "Telangana": "Telangana - Southern Cultural Region",
    "Puducherry": "Puducherry - Southern Cultural Region",
    "Lakshadweep": "Lakshadweep - Southern Cultural Region",

    "Gujarat": "Gujarat - Western Cultural Region",
    "Maharashtra": "Maharashtra - Western Cultural Region",
    "Goa": "Goa - Western Cultural Region",
    "Rajasthan": "Rajasthan - Western Cultural Region",
    "Dadra_and_Nagar_Haveli": "Dadra and Nagar Haveli - Western Cultural Region",
    "Daman_and_Diu": "Daman and Diu - Western Cultural Region",

    "West_Bengal": "West Bengal - Eastern Cultural Region",
    "Odisha": "Odisha - Eastern Cultural Region",
    "Bihar": "Bihar - Eastern Cultural Region",
    "Jharkhand": "Jharkhand - Eastern Cultural Region",
    "Andaman_and_Nicobar": "Andaman and Nicobar Islands - Eastern Cultural Region",

    "Madhya_Pradesh": "Madhya Pradesh - Central Cultural Region",
    "Chhattisgarh": "Chhattisgarh - Central Cultural Region",

    "Assam": "Assam - Northeastern Cultural Region",
    "Arunachal_Pradesh": "Arunachal Pradesh - Northeastern Cultural Region",
    "Manipur": "Manipur - Northeastern Cultural Region",
    "Meghalaya": "Meghalaya - Northeastern Cultural Region",
    "Mizoram": "Mizoram - Northeastern Cultural Region",
    "Nagaland": "Nagaland - Northeastern Cultural Region",
    "Sikkim": "Sikkim - Northeastern Cultural Region",
    "Tripura": "Tripura - Northeastern Cultural Region"
}

# --------------------------------------------------
# SETTINGS
# --------------------------------------------------

BATCH_SIZE = 40

# --------------------------------------------------
# PARSER
# --------------------------------------------------

def extract_sentiment(text):

    text = text.strip()

    try:
        match = re.search(r"\{.*?\}", text, re.DOTALL)

        if match:
            obj = json.loads(match.group())

            sentiment = (
                obj["sentiment"]
                .lower()
                .strip()
            )

            if sentiment in [
                "positive",
                "neutral",
                "negative"
            ]:
                return sentiment

    except Exception:
        pass

    text_lower = text.lower()

    if "positive" in text_lower:
        return "positive"

    if "negative" in text_lower:
        return "negative"

    if "neutral" in text_lower:
        return "neutral"

    return "PARSE_ERROR"

# --------------------------------------------------
# BATCH INFERENCE
# --------------------------------------------------

def classify_batch(texts, location):

    prompts = []

    for txt in texts:

        prompts.append(
            PROMPT_TEMPLATE.format(
                location=location,
                text=str(txt)
            )
        )

    inputs = tokenizer(
        prompts,
        return_tensors="pt",
        padding="longest",
        truncation=True,
        max_length=512
    ).to(model.device)

    with torch.no_grad():

        generated_outputs = model.generate(
            **inputs,
            max_new_tokens=30,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id
        )

    results = []

    for i in range(len(prompts)):

        input_len = (
            inputs["attention_mask"][i]
            .sum()
            .item()
        )

        generated = generated_outputs[i][input_len:]

        decoded = tokenizer.decode(
            generated,
            skip_special_tokens=True
        )

        results.append(
            extract_sentiment(decoded)
        )

    return results

# --------------------------------------------------
# MAIN LOOP
# --------------------------------------------------

for state_name, location in LOCATIONS.items():

    state_file = (
        f"state_results/{state_name}.csv"
    )

    if os.path.exists(state_file):

        print(
            f"Skipping {state_name} "
            "(already completed)"
        )

        continue

    print(
        f"\nProcessing {state_name}"
    )

    sentiments = []

    for start in tqdm(
        range(
            0,
            len(data),
            BATCH_SIZE
        )
    ):

        batch_texts = (
            data.iloc[
                start:
                start + BATCH_SIZE
            ]
            .astype(str)
            .tolist()
        )

        batch_predictions = classify_batch(
            batch_texts,
            location
        )

        sentiments.extend(
            batch_predictions
        )

    state_output = pd.DataFrame({
        "Hindi": data.astype(str),
        "State": state_name,
        "Region": location,
        "Sentiment": sentiments
    })

    state_output.to_csv(
        state_file,
        index=False,
        encoding="utf-8-sig"
    )

    print(
        f"Saved {state_file}"
    )

# --------------------------------------------------
# REGION FILES
# --------------------------------------------------

print(
    "\nCreating region files..."
)

region_tables = defaultdict(list)

for state_name, location in LOCATIONS.items():

    state_file = (
        f"state_results/{state_name}.csv"
    )

    if os.path.exists(state_file):

        df_state = pd.read_csv(
            state_file
        )

        region_tables[
            location
        ].append(df_state)

for region_name, dfs in region_tables.items():

    region_df = pd.concat(
        dfs,
        ignore_index=True
    )

    safe_region_name = (
        region_name
        .replace(" ", "_")
        .replace("-", "_")
    )

    region_file = (
        f"region_results/"
        f"{safe_region_name}.csv"
    )

    region_df.to_csv(
        region_file,
        index=False,
        encoding="utf-8-sig"
    )

    print(
        f"Saved {region_file}"
    )

print(
    "\nFinished successfully."
)
```
