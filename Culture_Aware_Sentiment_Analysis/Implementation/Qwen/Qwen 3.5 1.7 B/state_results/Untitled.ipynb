```python
import pandas as pd

rows = []

for (hindi, region), g in df.groupby(["Hindi", "Region"]):
    sample = (
        g.drop_duplicates(subset="State")
         .sample(n=min(3, g["State"].nunique()), random_state=42)
         .copy()
    )

    sample["Hindi"] = hindi
    sample["Region"] = region

    rows.append(sample)

qwendf = (
    pd.concat(rows, ignore_index=True)
      [["Hindi", "Region", "State", "Sentiment"]]
)

print(qwendf.head())
```

```python

qwendf["Region"] = qwendf["Region"].str.replace(
    r"^.*? - ",
    "",
    regex=True
)
```

```python
import pandas as pd

# Randomly select at most 3 states from each region
selected_states = (
    qwendf.groupby("Region")["State"]
           .apply(lambda s: s.drop_duplicates().sample(
               n=min(3, s.nunique()),
               random_state=42
           ).tolist())
           .to_dict()
)

# Filter qwendf
qwen_output_df = qwendf[
    qwendf.apply(
        lambda row: row["State"] in selected_states[row["Region"]],
        axis=1
    )
].reset_index(drop=True)

# Keep only the required columns
qwen_output_df = qwen_output_df[
    ["Hindi", "Region", "State", "Sentiment"]
]
```
