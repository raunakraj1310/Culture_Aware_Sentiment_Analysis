```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
anls.plot(kind='bar', figsize=(10, 6))

plt.xlabel("Sentiments")
plt.ylabel("Count of Sentiment")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title="Models")
plt.tight_layout()
plt.title("Sentiment Distribution Across Models")
plt.show()

```

```python
labels=["Negative","Neutral","Positive"]
colors=["red","yellow","green"]
sizes=[2411,5270,4151]
plt.figure(figsize=(8,10))
# plt.figure(figsize=(7,7))
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct="%1.00001f%%",
    startangle=90,
    counterclock=False

)
plt.title("Sentiment Distribution for Gemma_4B")
plt.axis("equal")
plt.legend()
plt.show()
```

```python
counts=sentiment_df["Gemma_1B"].value_counts()
labels=counts.index
sizes=counts.values
col_map={
    "PARSE_ERROR": "black",
    "negative": "red",
    "neutral": "yellow",
    "positive": "green"
}
colors=[col_map[label] for label in labels]
plt.figure(figsize=(8,10))
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct="%1.1f%%"
)
plt.title("Sentiment Distribution for Gemma_1B")
plt.axis()
plt.legend(title="Sentiments")
plt.show()



# print(counts,labels,sizes,colors,sep="\n")
```
